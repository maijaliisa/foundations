```python
history_df = pd.read_csv('History_Use_This_one.csv')
history_df
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
def make_point(row):
    return Point(row.Longitude, row.Latitude)
history_df['geometry'] = history_df.apply(make_point, axis=1)
```

```python
history_df.head()
```

```python
history_df = gpd.GeoDataFrame(history_df)
```

```python
history_df.to_file('Use This For History.json', driver='GeoJSON')
```
